```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import datetime
from datetime import datetime
```

```python
gdata = pd.read_csv("C:\data_files\GOOG.csv", usecols = ['Date','Open','Close'], index_col = 0)
gdata
```

```python
adata = pd.read_csv("C:\data_files\AMZN.csv", usecols = ['Date', 'Open', 'Close'], index_col = 0)
adata
```

```python
gdata['Gain'] = gdata['Close'] - gdata['Open']
gdata
```

```python
adata['Gain'] = adata['Close'] - adata['Open']
adata
```

```python
gdata['PctGain'] = (gdata['Gain'] / gdata['Open']) * 100
gdata
```

```python
adata['PctGain'] = (adata['Gain'] / adata['Open']) * 100
adata
```

```python
stockwin = gdata.join(adata, lsuffix = '_g', rsuffix = '_a')
stockwin
```

```python
stockwin['GoogleWins'] = (stockwin.PctGain_g >= stockwin.PctGain_a)
stockwin
```

```python
GWins = stockwin[stockwin.PctGain_g > stockwin.PctGain_a]
GWins
```

```python
GWinPct = GWins['GoogleWins'].count() / stockwin['GoogleWins'].count() * 100
GWinPct
```
